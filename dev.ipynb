```python
def calculate_cust_id(id1: str, id2: str) -> str:
    """
    當 cust_id 為空時，依 id1 及 id2 計算得到 cust_id 欄位
    id2 取最末 2 碼，id1 取完整 8 碼組成電號前 10 碼，最後一碼檢核碼依查表後產生
    """
    id1 = id1.zfill(8)
    id2 = id2.zfill(2)
    parts = [id2[-2:], id1[:2], id1[2:4], id1[4:6], id1[6:8]]
    nums = [int(20) for part in parts]
    print(parts)
    checkval = sum(nums) % 10
    return id2[-2:] + id1 + str(checkval)

calculate_cust_id("87654321", "12")
calculate_cust_id("12345678", "12")
```
